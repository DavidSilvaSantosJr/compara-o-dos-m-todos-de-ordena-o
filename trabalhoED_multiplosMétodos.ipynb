# iniciar ordenação


```python
from matplotlib.pyplot import legend
from numpy import arange
from random import randint, shuffle
import pylab
import time
from datetime import datetime

qtdValores_ordenar = [100, 1000, 5000, 10000, 20000, 100000, 200000, 500000]

#alimentar listas
def gerarEadicionar(vetor):
    for c in range(len(vetor)):
        for v in range(qtdValores_ordenar[c]):
            vetor[c].append(randint(0, 50))
    return vetor

#metodos ordenação
#bolha
def ordenar_bubbleSort(v):
    inicio = datetime.now()
    global somarInicioContagem
    somarInicioContagem = datetime.timestamp(inicio)

    for passnum in range(len(v)-1,0,-1):
        for i in range(passnum):
            if v[i] > v[i+1]:
                temp = v[i]
                v[i] = v[i+1]
                v[i+1] = temp

    fim = datetime.now()
    global somarFimContagem
    somarFimContagem = datetime.timestamp(fim)

    return ordenar_bubbleSort

#cocktail
def ordenar_cocktailSort(listaValores):
    i = datetime.now()
    global somarInicioContagem
    somarInicioContagem = datetime.timestamp(i)

    n = len(listaValores)
    flag = True
    start = 0
    end = n-1
    while (flag==True):
        # to ignore the result of the previous iteration
        flag = False
        # left to right traversal
        for i in range (start, end):
            if (listaValores[i] > listaValores[i+1]) :
                listaValores[i], listaValores[i+1]=listaValores[i+1], listaValores[i]
            flag=True
        # if no swap takes place array remains sorted
        if (flag==False):
            break
        # otherwise, reset the flag
        flag = False
        # last item is aldready sorted
        end = end-1
        # iteration from right to left
        for i in range(end-1, start-1,-1):
            if (listaValores[i] > listaValores[i+1]):
                listaValores[i], listaValores[i+1] = listaValores[i+1], listaValores[i]
            flag = True
        # first element is already sorted
        start = start+1

    f = datetime.now()
    global somarFimContagem
    somarFimContagem = datetime.timestamp(f)   
    return listaValores

#bucket
def ordenar_bucketSort(listaValores):
    for v in listaValores:
        i = datetime.now()
        global somarInicioContagem
        somarInicioContagem = datetime.timestamp(i)

        largest = max(listaValores)
        length = len(listaValores)
        size = largest/length
        # criar buckets
        buckets = [[] for i in range(length)]
        # ordenar buckets   
        for i in range(length):
            index = int(listaValores[i]/size)
            if index != length:
                buckets[index].append(listaValores[i])
            else:
                buckets[length - 1].append(listaValores[i])
        # ordnar buckets individualmente
        for i in range(len(listaValores)):
            buckets[i] = sorted(buckets[i]) 
        # achatar a lista
        global result
        result = []
        for i in range(length):
            result = result + buckets[i]

        f = datetime.now()
        global somarFimContagem
        somarFimContagem = datetime.timestamp(f)        
    return result

#merge
def ordenar_mergeSort(listaValores):

    i = datetime.now()
    global somarInicioContagem
    somarInicioContagem = datetime.timestamp(i)
    # 1. Armazenar o cumprimento da lista

    list_length = len(listaValores)

    # 2. Lista com cumprimento menor do que já está ordenado
    if list_length == 1:
        return listaValores

    # 3.Identificar o médio da lista e particione a lista em uma partição_esquerda e uma partição_direita
    mid_point = list_length // 2


    # 4. Para garantir que todas as partições sejam seguras em seus componentes individuais,
    # a função merge_sort é chamada e uma parte particionada da lista é tardia como parâmetro
    left_partition = ordenar_mergeSort(listaValores[:mid_point])
    right_partition = ordenar_mergeSort(listaValores[mid_point:])


    # 5. A função mesclar uma lista composta por uma classificação ordenada retorna à esquerda e à direita.
    return merge(left_partition, right_partition)
# 6. recebe duas listas e devolve uma lista ordenada composta pelo conteúdo das duas listas

def merge(left, right):
    # 7. Inicialização de uma lista vazia que terá uma saída com itens selecionados.
    # Inicializa duas variáveis ​​iej que são ponteiros usados ​​ao percorrer as listas.
    output = []
    i = j = 0

    # 8. Executa o loop enquanto ambos os ponteiros são menores que o comprimento das listas esquerda e direita
    while i < len(left) and j < len(right):
        # 9. Compare os elementos em cada posição de ambas as listas durante cada iteração
        if left[i] < right[j]:
            # a viagem é com o menor valor
            output.append(left[i])
            # 10. Mover o ponteiro para a direita
            i += 1
        else:
            output.append(right[j])
            j += 1
    # 11. Os elementos remanescentes são selecionados do valor atual do ponteiro até o final da lista respectivamente
    output.extend(left[i:])
    output.extend(right[j:])

    f = datetime.now()
    global somarFimContagem
    somarFimContagem = datetime.timestamp(f)

    return output

#quick
def partition(arr, low, high):
    i = datetime.now()
    global somarInicioContagem
    somarInicioContagem = datetime.timestamp(i)

    iQuick = (low-1)         # índice do elemento menor
    pivot = arr[high]     # pivô
  
    for j in range(low, high):
  
        # Se o elemento atual for menor ou
        # igual a pivô
        if arr[j] <= pivot:
  
            # índice de incremento do elemento menor
            iQuick = iQuick+1
            arr[iQuick], arr[j] = arr[j], arr[iQuick]
    
    
  
    arr[iQuick+1], arr[high] = arr[high], arr[iQuick+1]
    return (iQuick+1)
  
# A principal função que implementa o QuickSort
# arr[] --> Matriz a ser classificada,
# low  --> Índice inicial,
# high  --> Índice final
  
# Função para Quick sort
def ordenar_quickSort(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:
  
        # pi é o índice de particionamento, arr[p] is now
        # at right  
        pi = partition(arr, low, high)
  
        # Classifique os elementos separadamente antes
        # partição e depois da partição
        ordenar_quickSort(arr, low, pi-1)
        ordenar_quickSort(arr, pi+1, high)

    f = datetime.now()
    global somarFimContagem
    somarFimContagem = datetime.timestamp(f)

#shell sort
def ordenar_shellSort(array, n):
    i = datetime.now()
    global somarInicioContagem
    somarInicioContagem = datetime.timestamp(i)
    # Rearrange elements at each n/2, n/4, n/8, ... intervals
    interval = n // 2
    while interval > 0:
        for ishell in range(interval, n):
            temp = array[ishell]
            j = ishell
            while j >= interval and array[j - interval] > temp:
                array[j] = array[j - interval]
                j -= interval

            array[j] = temp
        interval //= 2

    f = datetime.now()
    global somarFimContagem
    somarFimContagem = datetime.timestamp(f)

#perm
def ordenar_permSort(listaValores):
    iTempo = datetime.now()
    global somarInicioContagem
    somarInicioContagem = datetime.timestamp(iTempo)

    def is_sorted(listaValores): #função vê se vetor está ordenado
        for i in range(1,len(listaValores)):
            if listaValores[i]<listaValores[i - 1]: return False
        return True

    def perm_sort(listaValores): #define perm sort
        ct=0
        while not is_sorted(listaValores): # caso o vetor não esteja ordenado após o perm, repetir o ciclo.
            shuffle(listaValores)
            ct+=1
        return ct, listaValores

    f = datetime.now()
    global somarFimContagem
    somarFimContagem = datetime.timestamp(f)
    return listaValores

valoresBolha = [list(), list(),list(),list(),list(),list(),list(),list()]
valoresCocktail = [list(), list(),list(),list(),list(),list(),list(),list()]
valoresBucket = [list(), list(),list(),list(),list(),list(),list(),list()]
valoresMerge = [list(), list(),list(),list(),list(),list(),list(),list()]
valoresQuick = [list(), list(),list(),list(),list(),list(),list(),list()]
valoresShell = [list(), list(),list(),list(),list(),list(),list(),list()]
valoresPerm = [list(), list(),list(),list(),list(),list(),list(),list()]

tempoOrdenação_bolha = list()
tempoOrdenação_cocktail = list()
tempoOrdenação_bucket = list()
tempoOrdenação_merge = list()
tempoOrdenação_quick = list()
tempoOrdenação_shell = list()
tempoOrdenação_perm = list()

#bubble sort
print(f'{">>>bubble.sort()<<<"}')
print(f'{"valores ordenados":>15}{"tempo ordenação (s)":>33}')
gerarEadicionar(valoresBolha)
for b in range(len(valoresBolha)):
    ordenar_bubbleSort(valoresBolha[b])
    tempoOrdenação_bolha.append(somarFimContagem - somarInicioContagem)
    print(f'{qtdValores_ordenar[b]} valores ordenados em {tempoOrdenação_bolha[b]:_>30} s')
print('='*50)

#cocktail sort
print(f'{">>cocktail.sort()<<<"}')
print(f'{"valores ordenados":>15}{"tempo ordenação (s)":>33}')
gerarEadicionar(valoresCocktail)
for c in range(len(valoresCocktail)):
    ordenar_cocktailSort(valoresCocktail[c])
    tempoOrdenação_cocktail.append(somarFimContagem - somarInicioContagem)
    print(f'{qtdValores_ordenar[c]} valores ordenados em {tempoOrdenação_cocktail[c]:_>40} s')
print('='*50)

#bucket sort
print(f'{">>bucket.sort()<<<"}')
print(f'{"valores ordenados":>15}{"tempo ordenação (s)":>33}')
gerarEadicionar(valoresBucket)
for bk in range(len(valoresBucket)):
    ordenar_bucketSort(valoresBucket[bk])
    tempoOrdenação_bucket.append(somarFimContagem - somarInicioContagem)
    print(f'{qtdValores_ordenar[bk]} valores ordenados em {tempoOrdenação_bucket[bk]:_>40} s')
print('='*50)

#merge sort
print(f'{">>>merge.sort()<<<"}')
print(f'{"valores ordenados":>15}{"tempo ordenação (s)":>33}')
gerarEadicionar(valoresMerge)
for m in range(len(valoresMerge)):
    ordenar_mergeSort(valoresMerge[m])
    tempoOrdenação_merge.append(somarFimContagem - somarInicioContagem)
    print(f'{qtdValores_ordenar[m]} valores ordenados em {tempoOrdenação_merge[m]:_>40} s')
print('='*50)

#quick sort
print(f'{">>>quick.sort()<<<"}')
print(f'{"valores ordenados":>15}{"tempo ordenação (s)":>33}')
gerarEadicionar(valoresQuick)
for q in range(len(valoresQuick)):
    ordenar_quickSort(valoresQuick[q], 0, len(valoresQuick) + 1)
    tempoOrdenação_quick.append(somarFimContagem - somarInicioContagem)
    print(f'{qtdValores_ordenar[q]} valores ordenados em {tempoOrdenação_quick[q]:_>40} s')
print('='*50)

#shell sort
print(f'{">>>shell.sort()<<<"}')
print(f'{"valores ordenados":>15}{"tempo ordenação (s)":>33}')
gerarEadicionar(valoresShell)
for s in range(len(valoresShell)):
    ordenar_shellSort(valoresShell[s], len(valoresShell[s]))
    tempoOrdenação_shell.append(somarFimContagem - somarInicioContagem)
    print(f'{qtdValores_ordenar[s]} valores ordenados em {tempoOrdenação_shell[s]:_>40} s')
print('='*50)

#perm sort
print(f'{">>>perm.sort()<<<"}')
print(f'{"valores ordenados":>15}{"tempo ordenação (s)":>33}')
gerarEadicionar(valoresPerm)
for p in range(len(valoresPerm)):
    ordenar_permSort(valoresPerm[p])
    tempoOrdenação_perm.append(somarFimContagem - somarInicioContagem)
    print(f'{qtdValores_ordenar[p]} valores ordenados em {tempoOrdenação_perm[p]:_>40} s')
print('='*50)
```
